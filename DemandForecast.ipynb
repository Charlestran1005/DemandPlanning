```python
import pandas as pd
from lunarcalendar.converter import Lunar, Solar, Converter
import datetime

# Danh sách các ngày lễ cố định (tháng-ngày Dương lịch) - Giỗ Tổ đã bị loại bỏ
fixed_holidays_dm = [ # dm means day-month, still used to generate dates
    '01-01',  # Tết Dương Lịch
    '30-04',  # Ngày Giải phóng Miền Nam
    '01-05',  # Ngày Quốc tế Lao động
    '02-09',  # Ngày Quốc khánh
]

def get_tet_holiday_range_dates(year, tet_start_date_str, tet_end_date_str):
    """Trả về danh sách ngày nghỉ Tết (datetime.date objects)."""
    start_date = pd.to_datetime(tet_start_date_str).date()
    end_date = pd.to_datetime(tet_end_date_str).date()
    tet_dates = []
    current_date = start_date
    while current_date <= end_date:
        tet_dates.append(current_date) # Append datetime.date object directly
        current_date += datetime.timedelta(days=1)
    return tet_dates

def get_gio_to_hung_vuong_dates_dates(start_year, end_year):
    """Trả về danh sách ngày Giỗ Tổ (datetime.date objects)."""
    gio_to_dates = []
    for year in range(start_year, end_year + 1):
        gio_to_lunar = Lunar(year, 3, 10)
        gio_to_solar = Converter.Lunar2Solar(gio_to_lunar)
        gio_to_dates.append(gio_to_solar.to_date()) # Append datetime.date object
    return gio_to_dates

# --- NGÀY NGHỈ TẾT NGUYÊN ĐÁN (DƯƠNG LỊCH) 2022-2025 ---
tet_holiday_2022_start = '2022-01-29'
tet_holiday_2022_end = '2022-02-06'
tet_holiday_2023_start = '2023-01-20'
tet_holiday_2023_end = '2023-01-26'
tet_holiday_2024_start = '2024-02-08'
tet_holiday_2024_end = '2024-02-14'
tet_holiday_2025_start = '2025-01-28'
tet_holiday_2025_end = '2025-02-05'

# Tạo danh sách ngày nghỉ Tết và Giỗ Tổ (datetime.date objects)
tet_holidays_dates = []
for year in range(2022, 2026):
    if year == 2022:
        tet_holidays_dates.extend(get_tet_holiday_range_dates(year, tet_holiday_2022_start, tet_holiday_2022_end))
    elif year == 2023:
        tet_holidays_dates.extend(get_tet_holiday_range_dates(year, tet_holiday_2023_start, tet_holiday_2023_end))
    elif year == 2024:
        tet_holidays_dates.extend(get_tet_holiday_range_dates(year, tet_holiday_2024_start, tet_holiday_2024_end))
    elif year == 2025:
        tet_holidays_dates.extend(get_tet_holiday_range_dates(year, tet_holiday_2025_start, tet_holiday_2025_end))

gio_to_hung_vuong_holidays_dates = get_gio_to_hung_vuong_dates_dates(2022, 2025)

# Convert fixed holidays to datetime.date for each year in range
fixed_holidays_dates = []
for year in range(2022, 2026): # Adjust range if needed
    for holiday_dm in fixed_holidays_dm:
        day, month = map(int, holiday_dm.split('-'))
        try:
            holiday_date = datetime.date(year, month, day) # Create datetime.date for each year
            fixed_holidays_dates.append(holiday_date)
        except ValueError: # Handle cases like Feb 29th in non-leap year if needed, but not relevant here for fixed holidays
            pass # Or add logging/error handling

holidays_date_set = set(fixed_holidays_dates + tet_holidays_dates + gio_to_hung_vuong_holidays_dates) # Set of datetime.date objects
# --- CHECK 'FulfillDate' CÓ PHẢI NGÀY LỄ KHÔNG - VECTORIZED ---
sales['is_holiday_vectorized'] = sales['FulfillDate'].dt.date.isin(holidays_date_set) # Vectorized isin check
sales = sales.rename(columns={'is_holiday_vectorized': 'holiday'})

# Hiển thị kết quả DataFrame (chỉ vài dòng đầu)
sales.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error

# 🧹 Bước 1: Chuẩn hóa ngày & tạo cột tháng
am1182['OrderDate'] = pd.to_datetime(am1182['OrderDate'])
am1182['Month'] = am1182['OrderDate'].dt.to_period('M').astype(str)  # Ví dụ: '2024-01'

# 📊 Bước 2: Tổng hợp theo tháng (sum hoặc mean đều được)
df_monthly = am1182.groupby('Month')['OrderQuantity'].sum().reset_index()
df_monthly['Month'] = pd.to_datetime(df_monthly['Month'])

# 🔁 Bước 3: Tạo Naive Forecast = tháng trước đó
df_monthly['Naive_Forecast'] = df_monthly['OrderQuantity'].shift(1)

# 📈 Bước 4: Tính MAPE
df_eval = df_monthly.dropna()
mape = mean_absolute_percentage_error(df_eval['OrderQuantity'], df_eval['Naive_Forecast']) * 100
print(f"✅ MAPE (Naive Forecast theo tháng): {round(mape, 2)}%")

# 📈 Bước 5: Vẽ biểu đồ
plt.figure(figsize=(12,6))
plt.plot(df_monthly['Month'], df_monthly['OrderQuantity'], label='Thực tế (Tổng theo tháng)', marker='o')
plt.plot(df_monthly['Month'], df_monthly['Naive_Forecast'], label='Dự báo Naive (tháng trước)', linestyle='--', marker='x')
plt.title('Naive Forecast theo tháng - SKU AM1182')
plt.xlabel('Tháng')
plt.ylabel('Tổng số lượng đặt hàng')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

actual = df_eval['OrderQuantity']
forecast = df_eval['Naive_Forecast']

mae = mean_absolute_error(actual, forecast)
rmse = np.sqrt(mean_squared_error(actual, forecast))
bias = (forecast - actual).mean()
smape = (np.abs(forecast - actual) / ((np.abs(actual) + np.abs(forecast)) / 2)).mean() * 100
wape = np.sum(np.abs(forecast - actual)) / np.sum(actual) * 100

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"Bias: {bias:.2f}")
print(f"SMAPE: {smape:.2f}%")
print(f"WAPE: {wape:.2f}%")
```
